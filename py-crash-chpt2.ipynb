## Strings

```python
name="ada lovelace"
print(name.title())#he dot (.)after name in name.title() tells Python to make the title() method act on the variable name. The title() method changes each word to title case, where each word begins with a capital letter.
print(name.upper())
print(name.lower())
```

## f-strings

```python
first_name="ADA"
last_name="lovelace"
full_name=f"{first_name} {last_name}" 
#To insert a variable’s value into a string, place the letter f immediately before the opening quotation mark
#Put braces around the name or names of any variable you want to use inside the string. Python will replace each variable with its value when the string is displayed.
#These strings are called f-strings. The f is for format, because Python formats the string by replacing the name of any variable in braces with its value
print(f"Hello, {full_name.title()}!")
```

```python
first_name = "ada"
last_name = "lovelace"
full_name = f"{first_name} {last_name}"
message = f"Hello, {full_name.title()}!"
print(message)
```

## Whitespace

```python
print("Python")
print("\tPython")
print("Languages : \nPython\nC\nJavascript")
print("Languages:\n\tPython\n\tC\n\tJavaScript")
```

```python
#To ensure that no whitespace exists at the right side of a string, use the rstrip() method
favorite_language='python '
favorite_language

favorite_language.rstrip()
favorite_language
#However, it is only removed temporarily. If you ask for the valuebof favorite_language again, the string looks the same as when it was entered, including the extra whitespace 4 
favorite_language
```

```python
#To remove the whitespace from the string permanently, you have to associate the stripped value with the variable name
favorite_language='python '
favorite_language=favorite_language.rstrip()
favorite_language
```
